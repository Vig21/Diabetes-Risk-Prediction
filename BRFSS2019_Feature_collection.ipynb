```python
# Diabetes
# going to make this ordinal. 0 is for no diabetes or only during pregnancy, 1 is for pre-diabetes or borderline diabetes, 2 is for yes diabetes
# Remove all 7 (dont knows)
# Remove all 9 (refused)
# Remove NaNs
brfss2019['Diabetes'] = brfss2019['Diabetes'].replace({2:0, 3:0, 1:2, 4:1})
brfss2019 = brfss2019[brfss2019.Diabetes != 7]
brfss2019 = brfss2019[brfss2019.Diabetes != 9]
brfss2019 = brfss2019.dropna(subset = ['Diabetes'])
brfss2019.Diabetes.value_counts()
```

```python
# 1 HighBloodPressure
#Change 1 to 0 so it represetnts No high blood pressure and 2 to 1 so it represents high blood pressure
#Convert 9 (Don't know) to None
brfss2019['HighBloodPressure'].replace({1:0, 2:1, 9:None},inplace=True)
brfss2019.HighBloodPressure.value_counts(dropna=False)
```

```python
# 2 HighCholesterol
# Change 2 to 0 because it is No
# Convert all 7 (dont knows) to Nan
# Convert all 9 (refused) to Nan
brfss2019['HighCholesterol'] = brfss2019['HighCholesterol'].replace({2:0,7:None,9:None})
brfss2019.HighCholesterol.value_counts(dropna=False)
```

```python
# 3 CholesterolCheck
# 7 - Don't know/not sure
# 9 - Refused
# Convert 7,9 to None
brfss2019['CholesterolCheck'] = brfss2019['CholesterolCheck'].replace({7:None,9:None})
brfss2019.CholesterolCheck.value_counts(dropna = False)
```

```python
# 4 BodyMassIndex (no changes, just note that these are BMI * 100. So for example a BMI of 4018 is really 40.18)
brfss2019['BodyMassIndex'] = brfss2019['BodyMassIndex'].div(100)
print(brfss2019.BodyMassIndex.describe())
brfss2019.BodyMassIndex.isna().sum()
```

```python
# 5 Smoker
# Change 2 to 0 because it is No
# Remove all 7 (dont knows)
# Remove all 9 (refused)
brfss2019['Smoker'] = brfss2019['Smoker'].replace({2:0,7:None,9:None})
brfss2019.Smoker.value_counts(dropna=False)
```

```python
# 6 HadStroke
# Change 2 to 0 because it is No
# Remove all 7 (dont knows)
# Remove all 9 (refused)
brfss2019['HadStroke'] = brfss2019['HadStroke'].replace({2:0,7:None,9:None})
brfss2019.HadStroke.value_counts(dropna=False)
```

```python
# 7 HadHeartDiseaseorAtack
# Change 2 to 0 because this means did not have MI or CHD
brfss2019['HadHeartDiseaseorAtack'] = brfss2019['HadHeartDiseaseorAtack'].replace({2:0,7:None,9:None})
brfss2019.HadHeartDiseaseorAtack.value_counts(dropna=False)
```

```python
# 8 PhysicallyActive
# 1 for physical activity
# change 2 to 0 for no physical activity
# Convert all 9 (don't know/refused) to None
brfss2019['PhysicallyActive'] = brfss2019['PhysicallyActive'].replace({2:0,7:None,9:None})
# brfss2019 = brfss2019[brfss2019.PhysicallyActive != 9]
brfss2019.PhysicallyActive.value_counts(dropna=False)
```

```python
# 9 ConsumesFruits
# Change 2 to 0. this means no fruit consumed per day. 1 will mean consumed 1 or more pieces of fruit per day
# remove all dont knows and missing 9
brfss2019['ConsumesFruits'] = brfss2019['ConsumesFruits'].replace({2:0,7:None,9:None})
# brfss2019 = brfss2019[brfss2019.ConsumesFruits != 9]
brfss2019.ConsumesFruits.value_counts(dropna=False)
```

```python
# 10 ConsumesVeggies
# Change 2 to 0. this means no vegetables consumed per day. 1 will mean consumed 1 or more pieces of vegetable per day
# remove all dont knows and missing 9
brfss2019['ConsumesVeggies'] = brfss2019['ConsumesVeggies'].replace({2:0,7:None,9:None})
# brfss2019 = brfss2019[brfss2019.ConsumesVeggies != 9]
brfss2019.ConsumesVeggies.value_counts(dropna=False)
```

```python
# 11 HeavyAlcoholConsumption
# Change 1 to 0 (1 was no for heavy drinking). change all 2 to 1 (2 was yes for heavy drinking)
# convert all dont knows and missing 9 to None
brfss2019['HeavyAlcoholConsumption'] = brfss2019['HeavyAlcoholConsumption'].replace({1:0, 2:1, 9:None})
# brfss2019 = brfss2019[brfss2019.HeavyAlcoholConsumption != 9]
brfss2019.HeavyAlcoholConsumption.value_counts(dropna=False)
```

```python
# 12 HaveHealthCoverage
# 1 is yes, change 2 to 0 because it is No health care access
# Convert 7 and 9 for don't know or refused to Nan
brfss2019['HaveHealthCoverage'] = brfss2019['HaveHealthCoverage'].replace({2:0,7:None,9:None})
# brfss2019 = brfss2019[brfss2019.HaveHealthCoverage != 7]
# brfss2019 = brfss2019[brfss2019.HaveHealthCoverage != 9]
brfss2019.HaveHealthCoverage.value_counts(dropna=False)
```

```python
# 13 HaveHealthFinancialIssues
# Change 2 to 0 for no, 1 is already yes
# Convert 7 for don/t know and 9 for refused to Nan
brfss2019['HaveHealthFinancialIssues'] = brfss2019['HaveHealthFinancialIssues'].replace({2:0,7:None,9:None})
# brfss2019 = brfss2019[brfss2019.HaveHealthFinancialIssues != 7]
# brfss2019 = brfss2019[brfss2019.HaveHealthFinancialIssues != 9]
brfss2019.HaveHealthFinancialIssues.value_counts(dropna=False)
```

```python
# 14 GeneralHealthCondition
# This is an ordinal variable that I want to keep (1 is Excellent -> 5 is Poor)
# Convert 7 and 9 for don't know and refused to Nan
brfss2019['GeneralHealthCondition'] = brfss2019['GeneralHealthCondition'].replace({7:None,9:None})
# brfss2019 = brfss2019[brfss2019.GeneralHealthCondition != 7]
# brfss2019 = brfss2019[brfss2019.GeneralHealthCondition != 9]
brfss2019.GeneralHealthCondition.value_counts(dropna=False)
```

```python

# 15 MentalHealthCondition
# already in days so keep that, scale will be 0-30
# change 88 to 0 because it means none (no bad mental health days)
# convert 77 and 99 for don't know not sure and refused to Nan
brfss2019['MentalHealthCondition'] = brfss2019['MentalHealthCondition'].replace({88:0,77:None,99:None})
# brfss2019 = brfss2019[brfss2019.MentalHealthCondition != 77]
# brfss2019 = brfss2019[brfss2019.MentalHealthCondition != 99]
brfss2019.MentalHealthCondition.value_counts(dropna=False)
```

```python
# 16 PhysicalHealthCondition
# already in days so keep that, scale will be 0-30
# change 88 to 0 because it means none (no bad mental health days)
# Convert 77 and 99 for don't know not sure and refused to nan
brfss2019['PhysicalHealthCondition'] = brfss2019['PhysicalHealthCondition'].replace({88:0,77:None,99:None})
# brfss2019 = brfss2019[brfss2019.PhysicalHealthCondition != 77]
# brfss2019 = brfss2019[brfss2019.PhysicalHealthCondition != 99]
brfss2019.PhysicalHealthCondition.value_counts(dropna=False)
```

```python
# 17 DifficultyInWalking
# change 2 to 0 for no. 1 is already yes
# convert 7 and 9 for don't know not sure and refused to None
brfss2019['DifficultyInWalking'] = brfss2019['DifficultyInWalking'].replace({2:0,7:None,9:None})
# brfss2019 = brfss2019[brfss2019.DifficultyInWalking != 7]
# brfss2019 = brfss2019[brfss2019.DifficultyInWalking != 9]
brfss2019.DifficultyInWalking.value_counts(dropna=False)
```

```python
# 18 Gender
# in other words - is respondent male (somewhat arbitrarily chose this change because men are at higher risk for heart disease)
# change 2 to 0 (female as 0). Male is 1
brfss2019['Gender'] = brfss2019['Gender'].replace({2:0})
brfss2019.Gender.value_counts(dropna=False)
```

```python
# 19 AgeBand
# already ordinal. 1 is 18-24 all the way up to 13 wis 80 and older. 5 year increments.
# convert 14 to nan because it is don't know or missing
brfss2019['AgeBand'] = brfss2019['AgeBand'].replace({14:None})
# brfss2019 = brfss2019[brfss2019.AgeBand != 14]
brfss2019.AgeBand.value_counts(dropna=False)
```

```python
# 20 HighestLevelOfEducation
# This is already an ordinal variable with 1 being never attended school or kindergarten only up to 6 being college 4 years or more
# Scale here is 1-6
# Convert 9 for refused to None:
brfss2019['HighestLevelOfEducation'] = brfss2019['HighestLevelOfEducation'].replace({9:None})
# brfss2019 = brfss2019[brfss2019.HighestLevelOfEducation != 9]
brfss2019.HighestLevelOfEducation.value_counts(dropna=False)
```

```python
# 21 IncomeLevel
# Variable is already ordinal with 1 being less than $10,000 all the way up to 8 being $75,000 or more
# Convert 77 and 99 for don't know and refused to Nan
brfss2019['IncomeLevel'] = brfss2019['IncomeLevel'].replace({99:None,77:None})
# brfss2019 = brfss2019[brfss2019.IncomeLevel != 77]
# brfss2019 = brfss2019[brfss2019.IncomeLevel != 99]
brfss2019.IncomeLevel.value_counts(dropna=False)
```

```python
GROUP1_columns = ['wtchsalt','cholmed2','hombpchk','medshepb','dradvise','x.drdxar2','hadhyst2','hadmam','homrgchk','cvdinfr4','havarth4','pneuvac4','cvdcrhd4','x.rfchol2','bpmeds','chckdny2']
GROUP2_columns =['employ1']
GROUP3_columns =['joinpai2']
GROUP4_columns =['bphigh4','x.lmtwrk2','x.lmtact2','checkup1','cdassist']

for i in GROUP1_columns:
    brfss2019[i] = brfss2019[i].replace({2:0,7:None,9:None})
    print(brfss2019[i].value_counts(dropna=False))
for i in GROUP2_columns:
    brfss2019[i] = brfss2019[i].replace({9:None})
    print(brfss2019[i].value_counts(dropna=False))
for i in GROUP3_columns:
    brfss2019[i] = brfss2019[i].replace({77:None,99:None})
    print(brfss2019[i].value_counts(dropna=False))
for i in GROUP4_columns:
    brfss2019[i] = brfss2019[i].replace({7:None,9:None})
    print(brfss2019[i].value_counts(dropna=False))
```

```python
#SPECIAL CASES
#wtkg3
brfss2019['wtkg3'] = brfss2019['wtkg3'].div(100)
print(brfss2019.wtkg3.describe())
print(brfss2019.wtkg3.isna().sum())

#bmi category
print(brfss2019['x.bmi5cat'].value_counts(dropna=False))

#drop useless features
brfss2019  = brfss2019.drop(['weight2','colgsex','prediab1'],axis = 1)

# Handle alcday5 - standardizing the drinks
# Normalize to drinks per month
def normalize_drinks(alcday):
    if 100 <= alcday <= 199:  # Days per week
        days_per_week = alcday - 100
        return min(days_per_week * 4.33,30)  # Convert to days per month
    elif 200 <= alcday <= 299:  # Days per month
        return alcday - 200
    elif alcday == 888:  # No drinks in past 30 days
        return 0
    else:  # 777 or 999 (unsure or refused)
        return np.nan
brfss2019['alcday5'] = brfss2019['alcday5'].apply(normalize_drinks)
print(brfss2019.alcday5.describe())
print(brfss2019.alcday5.isna().sum())
```

Class imblance ratio and handle it

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=9010)
rf_classifier.fit(X_train, y_train)


y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
