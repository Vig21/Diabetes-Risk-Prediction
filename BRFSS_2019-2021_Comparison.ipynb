```python
import pandas as pd
```

```python
df2019 = pd.read_csv('/content/drive/MyDrive/DM_Fall24_Project/Cleaned_BRFSS2019.csv')
```

```python
df2021 = pd.read_csv('/content/drive/MyDrive/DM_Fall24_Project/Cleaned_BRFSS2021.csv')
```

```python
# For each column, get value_counts for both df2019 and df2021
for col in ['Diabetes', 'HighBloodPressure', 'HighCholesterol', 'CholesterolCheck', 'BodyMassIndex', 'Smoker',
            'HadStroke', 'HadHeartDiseaseorAtack', 'PhysicallyActive', 'ConsumesFruits', 'ConsumesVeggies',
            'HeavyAlcoholConsumption', 'HaveHealthCoverage', 'HaveHealthFinancialIssues', 'GeneralHealthCondition',
            'MentalHealthCondition', 'PhysicalHealthCondition', 'DifficultyInWalking', 'Gender', 'AgeBand',
            'HighestLevelOfEducation', 'IncomeLevel']:

    print(f"Value counts for column: {col}")
    print("df2019:")
    print(df2019[col].value_counts())
    print("df2021:")
    print(df2021[col].value_counts())
    print("\n" + "-"*40 + "\n")
```
