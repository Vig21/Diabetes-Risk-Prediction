```python
import pandas as pd
```

```python
df2021 = pd.read_csv('/content/drive/MyDrive/CSE572 Data Mining/Initial Dataset/BRFSS Data/brfss2021_cleaned.csv')
df2021.shape
```

```python
df2019 = pd.read_csv('/content/drive/MyDrive/CSE572 Data Mining/Initial Dataset/BRFSS Data/brfss2019_cleaned.csv')
df2019.shape
```

```python
combined_df = pd.concat([df2019, df2021])
combined_df.shape
```

```python
combined_df.sample(5)
```

```python
combined_df.Diabetes.value_counts()
```

```python
for col in combined_df.columns:
  print(col, combined_df[col].value_counts())
  print()
```

```python
combined_df = combined_df[combined_df.Diabetes != 1]
```

```python
# prompt: aggregate BMI column into 6 bins based on CDC's categories for BMI

bins = [0, 18.5, 25, 30, 35, 40, float('inf')]
labels = ['Underweight', 'Normal Weight', 'Overweight', 'Obesity Class 1', 'Obesity Class 2', 'Obesity Class 3']
labels_ordinal = [1, 2, 3, 4, 5, 6]
combined_df['BMI_Category'] = pd.cut(combined_df['BodyMassIndex'], bins=bins, labels=labels_ordinal, include_lowest=True)
```

```python
combined_df.BMI_Category.value_counts()
```

```python
# prompt: aggregate MentalHealthCondition column into 6 bins. for example 1.0 - 5.0 into category 1, 5.0 - 10.0 into category 2 so on

bins = [0, 5, 10, 15, 20, 25, 30.0]
labels = [1, 2, 3, 4, 5, 6]

combined_df['MentalHealthCondition_Category'] = pd.cut(combined_df['MentalHealthCondition'], bins=bins, labels=labels, include_lowest=True)
combined_df.MentalHealthCondition_Category.value_counts()
```

```python
# prompt: aggregate PhysicalHealthCondition column into 6 bins. for example 1.0 - 5.0 into category 1, 5.0 - 10.0 into category 2 so on

bins = [0, 5, 10, 15, 20, 25, 30]
labels = [1, 2, 3, 4, 5, 6]

combined_df['PhysicalHealthCondition_Category'] = pd.cut(combined_df['PhysicalHealthCondition'], bins=bins, labels=labels, include_lowest=True)
combined_df.PhysicalHealthCondition_Category.value_counts()
```

```python
# prompt: drop columns BodyMassIndex, MentalHealthCondition  and PhysicalHealthCondition

# Drop the specified columns
combined_df = combined_df.drop(['BodyMassIndex', 'MentalHealthCondition', 'PhysicalHealthCondition'], axis=1)
```

```python
for col in combined_df.columns:
  print(col, combined_df[col].value_counts())
  print()
```

```python
# prompt: convert all column values from float to int

for col in combined_df.columns:
  if pd.api.types.is_float_dtype(combined_df[col]):
    combined_df[col] = combined_df[col].fillna(0).astype(int)
```

```python
combined_df.to_csv('/content/drive/MyDrive/CSE572 Data Mining/Initial Dataset/BRFSS Data/clean_combined_df_binned.csv', index=False)
```
